```python
total = 0
import re
with open('day3.txt', 'r') as file:
    lines = file.read()
    matches = re.findall(r'mul\(\d+,\d+\)', lines)
    print(f"Matches: {matches}")
    for match in matches:
        print(f"Match: {match}")
        numbers = re.findall(r'\d+', match)
        print(f"Numbers: {numbers}")
        total = total + (int(numbers[0]) * int(numbers[1]))

print(f"Total: {total}")
```

```python
total = 0
import re
with open('day3.txt', 'r') as file:
    lines = file.read()
    matches = re.findall(r'mul\(\d+,\d+\)|do\(\)|don\'t\(\)', lines)
    enabled = True
    print(f"Matches: {matches}")
    for match in matches:
        print(f"Match: {match}")
        if match == "don't()":
            enabled = False
            print(f"Enabled: {enabled}")
        elif match == "do()":
            enabled = True
            print(f"Temp Match: {enabled}")
        elif enabled == False:
            print(f"Continue and skip this match because don't before")
            continue
        else:
            numbers = re.findall(r'\d+', match)
            print(f"Numbers: {numbers}")
            total = total + (int(numbers[0]) * int(numbers[1]))

print(f"Total: {total}")
```
