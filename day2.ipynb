```python
list_of_lines = []
list_of_unsafe = []

with open('day2.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        line_of_numbers = list(map(int, line.split()))
        list_of_lines.append(line_of_numbers)

for numbers in list_of_lines:
    number_temp = 0
    for i, number in enumerate(numbers):
        if i == 0:
            number_temp = number
        elif i == 1:
            is_increasing = number_temp - number
            is_decreasing = number_temp - number
            number_temp = number
        elif i > 1:
            if -3 <= is_increasing <= -1:
                is_increasing = number_temp - number
                if -3 <= is_increasing <= -1:
                    number_temp = number
                else:
                    list_of_unsafe.append(numbers)
                    break
            elif 1 <= is_decreasing <= 3:
                is_decreasing = number_temp - number
                if 1 <= is_decreasing <= 3:
                    number_temp = number
                else:
                    list_of_unsafe.append(numbers)
                    break
            else:
                list_of_unsafe.append(numbers)
                break

print(list_of_unsafe)
print(len(list_of_lines) - len(list_of_unsafe))


```

```python
removed_list = []
less_unsafe_numbers = []
unsafe_list = []

for numbers in list_of_unsafe:
    #print(f"numbers: {numbers}, removed: {removed_list}")
    removed_list.clear()
    for i in range(len(numbers)):
        temp_list = numbers[:]
        temp_list.pop(i)
        removed_list.append(temp_list)
        #print(f"i: {i}, numbers: {numbers}, temp: {temp_list}")
    unsafe_list.clear()
    for removed_numbers in removed_list:
        number_temp = 0
        #print(f"removed_numbers: {removed_numbers}")
        for i, number in enumerate(removed_numbers):
            #print(f"number: {number}")
            if i == 0:
                number_temp = number
            elif i == 1:
                is_order = number_temp - number
                number_temp = number
            elif i > 1:
                #print(f"order: {is_order}")
                if -3 <= is_order <= -1:
                    if -3 <= (number_temp - number) <= -1:
                        number_temp = number
                    else:
                        unsafe_list.append(number)
                        #print(f"break in increasing")
                        #print(f"unsafe_list: {unsafe_list}")
                        break
                elif 1 <= is_order <= 3:
                    if 1 <= (number_temp - number) <= 3:
                        number_temp = number
                    else:
                        unsafe_list.append(number)
                        #print(f"break in decreasing")
                        #print(f"unsafe_list: {unsafe_list}")
                        break
                else:
                    unsafe_list.append(number)
                    #print(f"break in equal or too high")
                    #print(f"unsafe_list: {unsafe_list}")
                    break
    if len(unsafe_list) == len(removed_list):
        less_unsafe_numbers.append(removed_list)
print(f"less_unsafe_numbers: {len(less_unsafe_numbers)}, save_numbers: {len(list_of_lines) - len(less_unsafe_numbers)}")
```
